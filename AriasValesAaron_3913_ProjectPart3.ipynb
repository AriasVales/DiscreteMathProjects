```python
import numpy as np

# Sample space and probabilities
sample_space = {"A":"treasure",  "B":"armor", "C":"weapons"}
probabilities = {"A": .4, "B": .5, "C": .1}


#Lets check the Axiom and make sure everything adds up like it should.
sum_of_probabilities = sum(probabilities.values())
print("Sum of probabilities:", sum_of_probabilities)


```

Alright let's go! I've got the sum of my probabilities to add up to 1! Lets continue on now that we know that axiom is satisfied.  Lets move on to creating a compound event and the probability. I gotta grind for some new weapons so let's see what we can find out with our information. 

```python
# Create a compound event by combining two simple events
compound_event = {"A or C": (sample_space["A"], sample_space["C"])}

# Calculate the probability of a union of events A and B
A_or_C_probability = probabilities["A"] + probabilities["C"] - (probabilities["A"] * probabilities["C"])
probabilities["A or C"] = A_or_B_probability

# Calculate the probability of an intersection of events A and B
A_and_C_probability = probabilities["A"] * probabilities["C"]
probabilities["A and C"] = A_and_B_probability

print("Compound event:", compound_event)
print("Probability of A or C:", A_or_C_probability)
print("Probability of A and C:", A_and_C_probability)
print("Updated probabilities:", probabilities)
```

What do all the crazy numbers and every thing mean?!?! Allow me to throw an explanation! Probability of a union of events: P(A  C) = P(A) + P(C) - P(A  C)
This formula calculates the probability of a union of two events A and C, where A  C represents the event that at least one of the events A or C occurs. The formula subtracts the probability of the intersection of events A and C (P(A  C)) from the sum of the probabilities of events A and C (P(A) + P(C)) to eliminate the double counting of the probability of both events occurring at the same time.

```python
# Conditional probability of A given C
P_A_given_C = probabilities["A"] / (probabilities["A"] + probabilities["C"])
print("P(A | C) =", P_A_given_C)

# Conditional probability of C given A
P_C_given_A = probabilities["C"] / (probabilities["A"] + probabilities["C"])
print("P(C | A) =", P_C_given_A)

# Bayes' theorem
P_A_given_C_verify = P_C_given_A * probabilities["A"] / probabilities["C"]
print("P(A | C) (verify) =", P_A_given_C_verify)

# Complement of A
P_notA = 1 - probabilities["A"]
print("P(not A) =", P_notA)

# Complement of C
P_notC = 1 - probabilities["C"]
print("P(not C) =", P_notC)
```

So lets explain what's going on above here! In this example, we first calculate the conditional probability of event A given that event C has already occurred, by dividing the probability of event A by the sum of the probabilities of events A and C. Next, we calculate the conditional probability of event C given that event A has already occurred, by dividing the probability of event C by the sum of the probabilities of events A and C. Finally, we apply Bayes' theorem to verify that the conditional probabilities are consistent with each other. For the complements we calculate the porbability of the complement of events A and C by subtracting their respective probabilities from 1! Easy right?

For my final act here I will Simulate a simple occurences of my sample space 10 times. I chose event "C" as my simulation to observe.  I need weapons to go forth and conqure dragons among other things. So I noticed that everytime I run this simulation it's different! Why? Because it's random! It's always going to come through slightly different every run through. This was alot of work and a ton of formulas to check out just the probability of what's gonna be in the box. 

```python
import random

sample_space = {"A": "treasure", "B": "armor", "C": "weapons"}
probabilities = [.4, .5, .1]

def simulate_sample_space():
    return random.choices(list(sample_space.values()), weights=probabilities, k=1)[0]

occurrences = [simulate_sample_space() for i in range(10)]
print("Occurrences:", occurrences)

# Count the number of times event C occurred
count_C = occurrences.count("weapons")
print("Number of times event C occurred:", count_C)

# Calculate the observed probability of event C
observed_prob_C = count_C / len(occurrences)
print("Observed probability of event C:", observed_prob_C)
```
