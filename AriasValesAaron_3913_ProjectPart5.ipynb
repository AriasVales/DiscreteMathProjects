For this example let's say my wife and I want a date night. This graph will show different locations such as home, theater, resturant, and concession stand.  The edges represent the time it takes for my wife and I to travel between locations and are weighted to represent the time in minutes it takes to get from location to another.  

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_edge("Home", "Movie Theater", weight=40)
G.add_edge("Home", "Restaurant", weight=40)
G.add_edge("Restaurant", "Movie Theater", weight=15)
G.add_edge("Movie Theater", "Concession Stand", weight=3)
G.add_edge("Concession Stand", "Movie Theater", weight=3)
G.add_edge("Movie Theater", "Restaurant", weight=15)
G.add_edge("Movie Theater", "Home", weight=40)
G.add_edge("Restaurant", "Home", weight=40)
nx.draw(G, with_labels=True)
plt.show()
```

##Pathways##
So pathway with the most options would be my wife and I going from home, to the resturant, to the movie theater, to the concession stand, to movie theater, then to home. The weights on that I have given are times between places.  The practicle use of this would be if my wife and I were trying to plan a baby sitter. So having all the times inbetween would help us better plan as well as ensuring the baby sitter had ample time. 
##Degrees of Vertecies##
So throughout this graph I kept it relatively simple. The Movie Theater has highest with three verticies because of the concession stand option. The concession stand is the lowest with one.  All other options have two.
